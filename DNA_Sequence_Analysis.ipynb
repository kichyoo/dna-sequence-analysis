```python
#import libraries 
from Bio import SeqIO #for reading sequences  
import pandas as pd #for data tables  
import matplotlib.pyplot as plt #for plotting  
```

```python
#load the file  
file_path=r"C:\Users\HP ELITEBOOK 830 G6\Documents\projects\project1\ecoli_genome.fasta"  

#read the first sequence in the file
record=next(SeqIO.parse(file_path,"fasta"))  
dna_sequence=str(record.seq) #convert sequence to a string  

#print basic info  
print(f"Sequence ID:{record.id}")  
print(f"Sequence length:{len(dna_sequence)}bases")  
```

```python
#calculate GC content  
gc_count=dna_sequence.count("G")+dna_sequence.count("C")  
total_bases=len(dna_sequence)  
gc_percent=(gc_count/total_bases)*100  

#print results  
print(f"GC Content:{gc_percent:.2f}%")  
```

```python
#count each nucleotide  
nucleotide_counts = {  
    "A": dna_sequence.count("A"),  
    "T": dna_sequence.count("T"),  
    "G": dna_sequence.count("G"),  
    "C": dna_sequence.count("C")  
}  

#create a bar plot  
plt.bar(nucleotide_counts.keys(),nucleotide_counts.values())  
plt.title("Nucleotide Frequency in E. coli Genome")  
plt.xlabel("Nucleotide")  
plt.ylabel("Count(millions)")  
plt.ylim(0,1.5e6) #adjust y-axis limit for readability  
plt.show()  
```

```python
#transcribe DNA to RNA  
rna_sequence=dna_sequence.replace("T","U")  

#print the first 50 bases of RNA  
print("First 50 RNA bases:",rna_sequence[:50])  
```
